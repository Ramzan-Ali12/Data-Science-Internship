```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
from PIL import Image

def load_image(image_path):
    image = Image.open(image_path)
    image = np.array(image)
    return image

def reshape_image(image):
    w, h, d = tuple(image.shape)
    image_reshaped = np.reshape(image, (w * h, d))
    return image_reshaped

def kmeans_clustering(image_reshaped, k):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(image_reshaped)
    labels = kmeans.predict(image_reshaped)
    cluster_centers = kmeans.cluster_centers_
    return labels, cluster_centers

def recreate_image(labels, cluster_centers, w, h, d):
    image_recreated = np.zeros((w, h, d))
    label_idx = 0
    for i in range(w):
        for j in range(h):
            image_recreated[i][j] = cluster_centers[labels[label_idx]]
            label_idx += 1
    return image_recreated.astype(np.uint8)

def plot_image(image, title):
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load the image
image_path = "/content/drive/MyDrive/Data Science Internship/week_2/Day_3/face.jpg"  # Replace with the path to your image
original_image = load_image(image_path)

# Reshape the image for K-Means clustering
reshaped_image = reshape_image(original_image)

# Perform K-Means clustering for different values of K
k_values = [2, 3, 5, 10, 15, 20]
for k in k_values:
    labels, cluster_centers = kmeans_clustering(reshaped_image, k)
    recreated_image = recreate_image(labels, cluster_centers, original_image.shape[0], original_image.shape[1], original_image.shape[2])
    plot_image(recreated_image, f"K={k}")

# Display the original image
plot_image(original_image, "Original Image")
```
